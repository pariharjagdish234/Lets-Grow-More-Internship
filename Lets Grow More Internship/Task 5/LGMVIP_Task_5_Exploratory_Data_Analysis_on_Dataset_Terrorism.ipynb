```python
# Renaming The Important Columns To Be Used
dataset=dataset.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','region_txt':'Region','provstate':'State','city':'City','attacktype1_txt':'Attack_type','targtype1_txt':'Target_type','targsubtype1_txt':'Target_subtype','gname':'Gang_name','weaptype1_txt':'Weapon_type'})
```

## Visualising & Analysing The Dataset

```python
# Observing The Most Used Weapon Type In Incidents of Attacks.
plt.figure(figsize=(20,10))
plt.xticks(rotation=45)
sb.countplot(x=dataset.Weapon_type)
plt.title("Weapon Type Usage")
```

**Observation: It is seen from the graph that Explosives followed by Firearms were the most frequent means of attack.**

```python
plt.figure(figsize=(20,10))
plt.xticks(rotation=45)
sb.countplot(x=dataset.Year)
plt.title("Years Of Attack")
```

**Observation: The terrorist activities had constant decreases from 1970 to 2013. But, it spiked in the years of 2014 & 2015.**

```python
# Observing The Terrorist Attacks By The Order Of Months In A Year
plt.figure(figsize=(20,10))
sb.countplot(x=dataset.Month)
plt.title("Months Of Attack")
```

**Observation: In almost all months of the year, the terrorist activity remained more or less the same.**

```python
plt.figure(figsize=(20,10))
sb.countplot(x=dataset.Day)
plt.title("Days Of Attack")
```

**Observation: The starting & ending of a month had least terrorist activity, while the terrorist activity was the highest just half past the month.**

```python
# Observing The Attacks With Respect To The Countries
plt.figure(figsize=(30,10))
plt.rcParams.update({'font.size':10})
plt.xticks(rotation=90)
sb.countplot(x=dataset.Country)
plt.title("Terrorist Activities In Various Countries")
```

**Observatiom: It is observed that Iraq dominates the terrorist activity chart,followed by Pakistan & Afghanistan.**

```python
# Observing The Attacks With Respect To Specific Regions
plt.figure(figsize=(30,10))
plt.rcParams.update({"font.size":20})
plt.xticks(rotation=45)
sb.countplot(x=dataset.Region)
plt.title("Regions Of Atttack")
```

**Observation: The regions of Middle East & Africa tops the chart followed closely by South Asia in terms of region specific terrorist attack.**

```python
plt.figure(figsize=(30,10))
plt.rcParams.update({'font.size':20})
plt.xticks(rotation=45)
sb.countplot(x=dataset.Attack_type)
plt.title("Nature Of The Attacks")
```

**Observation: Bombing/Explosion were the common nature of attack followed by armed assault.**

```python
plt.figure(figsize=(30,10))
plt.rcParams.update({'font.size':20})
plt.xticks(rotation=45)sb.countplot(x=dataset.Target_type)
plt.title("Attack On Target Types")
```

**Observation: The attacks were primarily focused on private citizens and property followed by the military groups.**

## Inferences From The Exploratory Analysis
